##1. Mean 

```python
import pandas as pd
import numpy as np

kolom = {'col1':[2,9,19],
         'col2':[3,np.nan,np.nan],
         'col3':[4,np.nan,18],
         'col4':[5,15,np.nan],
         'col5':[7,np.nan, 24]}

data = pd.DataFrame(kolom)
```

##4. Imputasi Frequent Category/Modus

```python
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np


mobil = {'merk':['ford', 'ford', 'toyota', np.nan, 'honda', 'yotota', 'ford', np.nan, 'ford', np.nan, 'honda', 'toyota', 'ford']}

data = pd.DataFrame(mobil)

data
```

```python
imp = SimpleImputer(strategy = 'most_frequent')
imp.fit_transform(data)
```
